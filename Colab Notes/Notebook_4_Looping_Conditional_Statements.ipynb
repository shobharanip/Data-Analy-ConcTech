#Looping Statements

1. For Loop

The for loop in Python is used to iterate over a sequence (like a list, tuple, dictionary, or string) and execute a block of code for each element in that sequence.

```python
# Example of a for loop
fruits = ['apple', 'banana', 'cherry']

for fruit in fruits:
    print(fruit)
```

Explanation:

    The for loop goes through each item in the fruits list one by one.
    For each iteration, the variable fruit takes the value of the current item, and the block of code inside the loop is executed.

```python
# for loops
# there are a few ways to go through a for loop
# one common method is to loop through a list
# it is also possible to loop through other types of aggregated variables
# but one rule remains the same
# the insides of a loop are determined by tabs!!!!!!!!
# this is also true for conditionals and user defined functions!!!!!!

counter = 0
simple_list = [0,1,2,3,4,5,6,2143,123,7834]
for i in simple_list:
    # the only lines that get executed in this loop are tabbed correctly!!!
    print(i)
    counter = counter + 1
print('this line is not inside the loop')
print('the loop cycled',counter,'times')
```

2. While Loop

The while loop repeats a block of code as long as a specified condition is True. Once the condition becomes False, the loop stops.

```python
# Example of a while loop
count = 0

while count < 3:
    print("Count is:", count)
    count += 1
```

Explanation:

    The loop continues as long as the value of count is less than 3.
    Each time the loop runs, it prints the current value of count and then increases count by 1.

## Conditionals

We will now look at conditional statements
There are 3 types of conditionals
1. if statement
2. else statement
3. elif statement

3. If Statement

The if statement is used to execute a block of code only if a specified condition is True.

```python
# Example of an if statement
number = 5

if number > 3:
    print("The number is greater than 3")
```

Explanation:

    The if statement checks whether the condition number > 3 is True.
    If the condition is True, the code inside the if block is executed.

4. If-Else Statement

The if-else statement is used to execute one block of code if a condition is True and another block if the condition is False.

```python
# Example of an if-else statement
number = 2

if number > 3:
    print("The number is greater than 3")
else:
    print("The number is not greater than 3")
```

Explanation:

    The if statement checks the condition number > 3.
    If the condition is True, the first block of code is executed; if it is False, the code inside the else block is executed.
    Since 2 is not greater than 3, this will output:

5. If-Elif-Else Statement

The if-elif-else statement is used when you have multiple conditions to check. The elif (short for "else if") allows you to check multiple conditions.

```python
# Example of an if-elif-else statement
number = 10

if number > 10:
    print("The number is greater than 10")
elif number == 10:
    print("The number is exactly 10")
else:
    print("The number is less than 10")
```
