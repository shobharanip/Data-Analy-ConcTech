# Data Dictionary
About the Dataset:
The dataset contains ten critical columns:
1. Temperature – This represents the average temperature recorded, which directly influences crop growth and yield.
2. Precipitation – Captures the amount of rainfall vital for soil moisture and crop production.
3. CO2 Levels – Denotes atmospheric CO2 concentration, which affects plant growth.
4. Crop Yield – Gauges the productivity of agricultural produce.
5. Soil Health – Indicates soil fertility and sustainability, essential for long-term agricultural success.
6. Extreme Weather Events – Represents occurrences like droughts or storms, negatively affecting crops.
7. Crop Disease Incidence – Measures the susceptibility of crops to diseases due to climate-related changes.
8. Water Availability – Shows the availability of water resources critical for irrigation.
9. Food Security – Reflects the stability and accessibility of food supplies.
10. Economic Impact – Highlights the financial implications of climate change on agriculture and related industries.

#Importance of Climate Change in Agriculture:
Climate change is a significant global challenge that profoundly impacts agriculture. Rising temperatures, changing precipitation patterns, and increasing CO2 levels are crucial in agricultural productivity. Extreme weather events such as droughts, floods, and storms disrupt farming operations, damaging crops and reducing yields. Changes in the ecosystem may increase the prevalence of crop diseases, while dwindling water resources can jeopardize irrigation, ultimately threatening food security.
As an environmental analyst, understanding these variables through data will allow you to evaluate how shifts in the climate can affect agricultural outcomes. By examining the data in this dataset, you will gain insights into the key factors that may drive changes in farming systems and help propose solutions to mitigate adverse effects.

## Univariate Analysis - Numerical Columns  -Histogram (for distribution of each numerical column)

```python
#4. Create a distribution plot for all numerical variables.
import seaborn as sns
import matplotlib.pyplot as plt

# Plot histograms for each numerical column
numerical_columns = df.select_dtypes(include=['number'])

# Plot histograms for each numerical column with a specific color
for column in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[column], kde=True, color='skyblue')  # Change color to 'skyblue'
    plt.title(f'Distribution of {column}')
    plt.show()
```

# Numerical Variables - Boxplot (for identifying outliers)

```python
# Plot boxplots for each numerical column with a custom palette
for column in numerical_columns:
    plt.figure(figsize=(8, 3))
    sns.boxplot(data=df, x=column, palette='Set3')  # Change palette to 'Set3'
    plt.title(f'Boxplot of {column}')
    plt.show()
```

# Numerical Variables - Violin *plot*

```python
# Plot violin plots for each numerical column with a custom palette
for column in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.violinplot(data=df, x=column, palette='BuPu')  # Change palette to 'Set2'
    plt.title(f'Violin Plot of {column}')
    plt.show()
```

#** 5.Count Plots (Categorical Variables)**

#5e. Univariate Analysis - Categorical Columns

```python
# 5. Create count plots for all categorical variables.
# Plot count plots for each categorical column with a custom palette and annotations
for column in categorical_columns:
    plt.figure(figsize=(12, 6))
    ax = sns.countplot(data=df, x=column, palette='Set2')  # Change palette to 'Set2'

    # Add count annotations on top of the bars
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.0f}',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center',
                    xytext=(0, 9),  # Adjust the position of the text
                    textcoords='offset points')

    plt.title(f'Count Plot of {column}')
    plt.xticks(rotation=45)
    plt.show()
```

#5f. Barplot with percentages

```python
# Define a custom color palette
custom_palette = sns.color_palette("Set2", len(categorical_columns))

# Plot bar plots with percentage for each categorical column
for i, column in enumerate(categorical_columns):
    plt.figure(figsize=(8, 4))
    cat_counts = df[column].value_counts(normalize=True) * 100
    sns.barplot(x=cat_counts.index, y=cat_counts.values,
                palette=[custom_palette[i % len(custom_palette)]])  # Rotate through the palette
    plt.title(f'Percentage Distribution of {column}')
    plt.ylabel('Percentage (%)')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Define a custom color palette
custom_palette = sns.color_palette("Set2", len(categorical_columns))

# Plot bar plots with percentage for each categorical column and annotate percentage numbers
for i, column in enumerate(categorical_columns):
    plt.figure(figsize=(10, 6))
    cat_counts = df[column].value_counts(normalize=True) * 100
    ax = sns.barplot(x=cat_counts.index, y=cat_counts.values,
                     palette=[custom_palette[i % len(custom_palette)]])  # Rotate through the palette

    # Add percentage annotations on top of the bars
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.1f}%',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center',
                    xytext=(0, 9),  # Adjust the position of the text
                    textcoords='offset points')

    plt.title(f'Percentage Distribution of {column}')
    plt.ylabel('Percentage (%)')
    plt.xticks(rotation=45)
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define a custom color palette
custom_palette = sns.color_palette("Set2")

# Plot count plots for each categorical column with normalized percentages
for column in categorical_columns:
    plt.figure(figsize=(12, 10))
    cat_counts = df[column].value_counts(normalize=True) * 100
    sns.barplot(x=cat_counts.index, y=cat_counts.values, palette=custom_palette)

    # Add percentage annotations
    for i, value in enumerate(cat_counts.values):
        plt.text(i, value + 1, f'{value:.1f}%', ha='center', va='bottom')

    plt.title(f'Percentage Distribution of {column}')
    plt.ylabel('Percentage (%)')
    plt.xticks(rotation=45)
    plt.show()
```

# **9. Box Plot (Extreme Weather vs Crop Yield)**

```python
# 9. What is the impact of extreme weather events on crop yield? Use a box plot to compare crop yields during high and low occurrences of extreme weather events.
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'Extreme Weather Events' is a categorical variable with values such as 'High' and 'Low',
# and 'Crop Yield' is the numerical variable.

# Box plot for 'Crop Yield' vs 'Extreme Weather Events'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Extreme Weather Events', y='Crop Yield', data=df, palette='Set2')
plt.title('Box Plot of Crop Yield by Extreme Weather Events')
plt.xlabel('Extreme Weather Events')
plt.ylabel('Crop Yield')
plt.xticks(rotation=45)
plt.show()
```
